```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm
from matplotlib.ticker import LinearLocator

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(0, 10, 0.25)
X, Y = np.meshgrid(X, Y)

# Define new parameters
a, b, c, d = -1, 1, 1, 0.05

E = c * X + d * Y * (0 - X)

Z = a * E + b * E**2

# Compute exposure function Z (reversing the original health function)
#Z = - (a * X - b * X**2 + c * Y - d * X * Y - e * X**2 * Y)

# Create figure and 3D axis
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Plot the surface with normalized axes
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Remove tick labels
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticklabels([])  # Hide Z-axis tick labels

# Add labels (reversed meaning)
ax.set_xlabel("Temperature (X)")
ax.set_ylabel("Adaptation / Income (Y)")
ax.set_zlabel("Dose-Response (Z)")

# Save figure
fig.savefig("../../output/figures/interaction_exposure_income.png", dpi=300, transparent=True)

# Show plot
plt.show()
```
